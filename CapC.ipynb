```python
import numpy as np
import pandas as pd
```

```python
#Preprocessing and Data Cleansing

df.iloc[0,:]

```

For the binary categorical variables:

Sex (F/M) will be encoded as 1 for Female (F) and 0 for Male (M).
Drinker (yes/no) will be encoded as 1 for Yes (drinker) and 0 for No (non-drinker).
For the target variable:

SMK_stat_type_cd, which represents the smoking state, will be encoded as follows:
0 for "never smoked" (SMK_stat_type_cd = 1)
1 for "has smoked or is currently a smoker" (SMK_stat_type_cd = 2 or 3)
This encoding simplifies the information, making it easier to work with these variables in a binary format.


```python
df["female"] = pd.get_dummies(df['sex'],1).iloc[:,0]

for i in range(df.shape[0]):
    if df['female'][i] == False:
        df['female'][i] = 0
    else:
        df['female'][i] = 1

df.iloc[:,[0,-1]]
```

```python
df["drinker"] = pd.get_dummies(df['DRK_YN'],0).iloc[:,1]
for i in range(df.shape[0]):
    if df['drinker'][i] == False:
        df['drinker'][i] = 0
    else:
        df['drinker'][i] = 1


df.iloc[:,[-3,-1]]
```

```python
df['smokers'] = df['SMK_stat_type_cd'] - 1 #sets ones to zero, twos to ones, and threes to twos
df['smokers'].loc[df['smokers'] > 1  ] = 1 #changes twos to ones
df['smokers'].describe() #mean shows us how many smokers in dataset
```

```python
nan_counts = [(column, df[column].isna().sum()) for column in df.columns]
nan_counts
```

```python
df.drop('SMK_stat_type_cd', axis=1, inplace=True)
df.drop("DRK_YN", axis=1, inplace=True)
df.drop("sex", axis=1, inplace =True)
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(16, 6))

# Calculate the correlation matrix using Pearson method
correlation_matrix = df.corr(method='pearson')

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='rainbow')

plt.title('Pairwise Correlation Heatmap', fontsize=12, pad=12)

# Show the plot
plt.show()
```
